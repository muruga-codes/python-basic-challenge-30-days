🚀 MINI PROJECT #2 — STUDENT RECORD MANAGER (CLI)

🧩 Problem Statement

Build a simple command-line Student Record Manager using Python.

🎯 FEATURES

Add student

View all students

Search by roll number

Delete student

Exit program

🪜 STEP-BY-STEP LOGIC

1️⃣ Use a list of dictionaries
2️⃣ Each student = {roll, name, marks}
3️⃣ Use while loop menu
4️⃣ Perform operations based on user choice

```python
student=[]
while True:
    print("\n 1.Add 2.View 3.Search 4.Delete 5.Exit")
    choice=input("enter your choice:")

    if choice=="1":
        rollno=input("Roll:")
        name=input("name:")
        marks=input("Marks:")
        student.append({"rollNO":rollno,"Name":name,"Mark":marks})
        print(student)
        print("STUDENTS ARE ADDED!!!")
    elif choice=="2":
        for s in student:
            print(s)
    elif choice=="3":
        r=input("enter a rollno to search:")
        for s in student:
            if s[rollno]==r:
                print(s)
    elif choice=="4":
        r=input("enter a rollno to delete:")
        for s in student:
            if s[rollno]!=r:
                print(s)
                print(student)
        print("ROll NO IS DELETED")
    elif choice=="5":
        break
```
