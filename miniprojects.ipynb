🚀 MINI PROJECT #2 — STUDENT RECORD MANAGER (CLI)

🧩 Problem Statement

Build a simple command-line Student Record Manager using Python.

🎯 FEATURES

Add student

View all students

Search by roll number

Delete student

Exit program

🪜 STEP-BY-STEP LOGIC

1️⃣ Use a list of dictionaries
2️⃣ Each student = {roll, name, marks}
3️⃣ Use while loop menu
4️⃣ Perform operations based on user choice

```python
student=[]
while True:
    print("\n 1.Add 2.View 3.Search 4.Delete 5.Exit")
    choice=input("enter your choice:")

    if choice=="1":
        rollno=input("Roll:")
        name=input("name:")
        marks=input("Marks:")
        student.append({"rollNO":rollno,"Name":name,"Mark":marks})
        print(student)
        print("STUDENTS ARE ADDED!!!")
    elif choice=="2":
        for s in student:
            print(s)
    elif choice=="3":
        r=input("enter a rollno to search:")
        for s in student:
            if s[rollno]==r:
                print(s)
    elif choice=="4":
        r=input("enter a rollno to delete:")
        for s in student:
            if s[rollno]!=r:
                print(s)
                print(student)
        print("ROll NO IS DELETED")
    elif choice=="5":
        break
```

🚀 MINI PROJECT — PASSWORD STRENGTH CHECKER

🎯 Project Goal
Build a Python program that checks whether a password is Weak, Medium, or Strong.

This type of logic is often asked in coding rounds and practical tests.

🧩 Requirements
The password must check:

Condition	Requirement
Length	    At least 8 characters
Uppercase	At least 1 uppercase letter
Lowercase	At least 1 lowercase letter
Digit	    At least 1 number
Specialchar At least 1 symbol (!@#$ etc.)

```python
def passwordchecker(password):
    if password is None:
        return None
    has_upper=False #initially set all into False
    has_lower=False
    has_digit=False
    has_special=False

    for ch in password:
        if ch.isupper():
            has_upper=True
        elif ch.islower():
            has_lower=True
        elif ch.isdigit():
            has_digit=True
        else:
            has_special=True
    score=sum([has_upper,has_lower,has_digit,has_special])

    if len(password)<8:
        return "weak password"
    elif score==4:
        return "Strong Password"
    elif score>=2:
        return "Medium Password"
    else:
        return "Strong Password"

password=input("Enter your Password:")
print(passwordchecker(password))
```
