Problem Statement

Create a program that analyzes a string and gives statistics.

FEATURES
Count vowels
Count consonants
Count words
Reverse string
Check palindrome

🪜 STEP BY STEP LOGIC
1️⃣ Take input from user
2️⃣ Loop through string
3️⃣ Count vowels using condition
4️⃣ Count words using split
5️⃣ Reverse string
6️⃣ Check palindrome

```python
def string_analye(text):
    if text is None:
        return None
    if not isinstance(text,str):
        print("give me only string")
        return
    s="aeiouAEIOU"
    vowels_count=0
    constant_count=0
    for ch in text:
        if ch in s:
            vowels_count+=1
        else:
            constant_count+=1
    return vowels_count,constant_count
vowels,constant=string_analye(text)  
words=len(text.split())
reverse=text[::-1]
palindrome=text==reverse

print("vowels:",vowels)
print("constant:",constant)
print("The count of the given words:",words)
print("The given string is palindrome:",palindrome)
print("The the reverse of the string is:",reverse)
```
