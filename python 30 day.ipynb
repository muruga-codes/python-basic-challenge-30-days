```python
def first_repeating(arr):
    if not arr:                 # edge case: empty list
        return None

    seen = set()

    for num in arr:
        if num in seen:
            return num          # first repeating element
        seen.add(num)

    return None                 # no repeating element


arr = [4, 5, 1, 2, 1, 4]
print(first_repeating(arr))
```

Write a program to find the maximum subarray sum (Kadane’s Algorithm). 
Input: [-2,1,-3,4,-1,2,1,-5,4] 
Output: 6

```python
def max_subarray_sum(arr):
    # Edge case: None input
    if arr is None:
        return None

    # Edge case: empty list
    if not arr:
        return 0

    max_current = arr[0]
    max_global = arr[0]

    for i in range(1, len(arr)):
        # Either extend previous subarray or start new from current element
        max_current = max(arr[i], max_current + arr[i])

        # Update global max if needed
        if max_current > max_global:
            max_global = max_current

    return max_global


arr = [-2,1,-3,4,-1,2,1,-5,4]
print(max_subarray_sum(arr))
```

 Write a program to find the longest consecutive sequence in a list.

Input: [100, 4, 200, 1, 3, 2]
Output: 4   (sequence: 1,2,3,4)


```python
def sequence(arr):
    if not arr:
        return[]
    nums=set(arr)
    long=0
    for num in nums:
        if num-1 not in nums:
            current=num
            length=1
        while current+1 in nums:
            current+=1
            length+=1
        long=max(long,length)
    return long
arr=[100, 4, 200, 1, 3, 2]
print(sequence(arr))
```

Write a program to check if two strings are isomorphic.

"egg" and "add" → True
"foo" and "bar" → False

```python
def is_isomorphic(s, t):
    # Edge case: None input
    if s is None or t is None:
        return False

    # Edge case: length mismatch
    if len(s) != len(t):
        return False

    map_st = {}
    map_ts = {}

    for i in range(len(s)):
        c1 = s[i]
        c2 = t[i]

        # Check mapping from s to t
        if c1 in map_st:
            if map_st[c1] != c2:
                return False
        else:
            map_st[c1] = c2

        # Check reverse mapping from t to s
        if c2 in map_ts:
            if map_ts[c2] != c1:
                return False
        else:
            map_ts[c2] = c1

    return True


# Tests
print(is_isomorphic("egg", "add"))  # True
print(is_isomorphic("foo", "bar"))  # False
```

approach 1:comparator with 'cmp_to_key'

```python
from functools import cmp_to_key

def largest_number(nums):
    strs = list(map(str, nums))

    def compare(a, b):
        if a + b > b + a:
            return -1 #high na -1
        elif a + b < b + a:
            return 1 #low na 1
        else:
            return 0 #equal na 0

    strs.sort(key=cmp_to_key(compare))

    if strs[0] == "0":
        return "0"

    return "".join(strs)
print(largest_number([3, 30, 34, 5, 9]))
```

validate an ip addresss

```python
def validate_ip(ip):
    parts = ip.split(".")
    
    if len(parts) != 4:
        return False
    
    for part in parts:
        if not part.isdigit():
            return False
        
        num = int(part)
        
        if num < 0 or num > 255:
            return False
        
        if part != str(num):
            return False
    
    return True


print(validate_ip("192.168.1.1"))   
print(validate_ip("256.1.1.1"))
```

Q1 — First non-repeating character in string

Example:
"aabbcde" → c

```python
def non_repeate(n):
    if n is None:
        return None
    freq={}
    for i in n:
        if i in freq:
            freq[i]+=1
        else:
            freq[i]=1
    for s in n:
        if freq[s]==1:
            return s

n="aabbcde"
print(non_repeate(n))
```

✅ Q1 — Find intersection of two arrays (without using set)
Example:
[1,2,2,3] and [2,2,4] → [2,2]

```python
def intersection(arr1, arr2):
    if not arr1 or not arr2:   # Edge case
        return []
    
    result = []
    arr2_copy = arr2[:]   # Make copy to avoid modifying original
    
    for num in arr1:
        if num in arr2_copy:
            result.append(num)
            arr2_copy.remove(num)   # Remove once matched
    
    return result


# Test
arr1 = [1,2,2,3]
arr2 = [2,2,4]
print(intersection(arr1, arr2))
```

✅ Q1 — Maximum Product of Two Elements in Array
Example:
[3,5,2,8,1] → 40
(5 * 8)



```python
def product(arr):
    if not arr:
        return []
    if len(arr)<2:
        return None
    large1=large2=float('-inf')
    small1=small2=float('inf')
    for num in arr:
        if num>large1:
            large2=large1
            large1=num
        elif num>large2:
            large2=num
        if num<small1:
            small2=small1
            small1=num
        elif num<small2:
            small2=num
    return max(large1*large2,small1*small2)
arr=[3,5,2,8,1]
print(product(arr))

```

✅ Q2 — Valid Parentheses (Stack Problem)
Input: "()[]{}" → True
Input: "(]" → False

(Very common coding round question)

```python
def parthesis(s):
    if s is None:
        return []
    stack=[]
    pairs={')':'(',']':'[','}':'{'}
    for ch in s:
        if ch in "([{":
            stack.append(ch)
        elif ch in ")]}":
            if not stack:
                return False
            if stack[-1]!=pairs[ch]:
                return False
            stack.pop()
    return len(stack)==0
print(parthesis("()[]{}"))
print(parthesis("(]"))
```

✅ Q4 — Longest Palindromic Substring (Basic version)
Input: "babad"Output: "bab" or "aba"
(Important interview logic)

```python
def palindrome(n):
    if n is None:
        return ""
    start=0
    end=0
    def expand(left,right):
        while left >=0 and right < len(n) and n[left]==n[right]:
            left-=1
            right+=1
        return left+1,right-1
    for i in range(len(n)):
        l1, r1 = expand(i, i)
        l2, r2 = expand(i, i + 1)

        if r1 - l1 > end - start:
            start, end = l1, r1

        if r2 - l2 > end - start:
            start, end = l2, r2

    return n[start:end+1]
#n="babad"
n="cbbd"
print(palindrome(n))
```

✅ Q1 — Longest Substring Without Repeating Characters
Example:

Input


"abcabcbb"
Output


3
Substring = "abc"

(Very common interview question)

```python
def longestsubstring(s):
    if s is None:
        return None
    Char_set=set()
    left=0
    max_length=0
    for right in range(len(s)):
        while s[right] in Char_set:
            Char_set.remove(s[left])
            left+=1
        Char_set.add(s[right])
        max_length=max(max_length,right-left+1)
    return max_length
s="abcabcbb"
print(longestsubstring(s))
```

✅ Q2 — Product of Array Except Self
Example:


Input: [1,2,3,4]
Output: [24,12,8,6]
Without using division.

(Very frequently asked)



```python
def productofarray(arr):
    if not arr:
        return []
    n=len(arr)
    result=[1]*n
    prefix=1
    for i in range(n):
        result[i]=prefix
        prefix*=arr[i]
    suffix=1
    for i in range(n-1,-1,-1):
        result[i]*=suffix
        suffix*=arr[i]
    return result
arr=[1,2,3,4]
print(productofarray(arr))
```

✅ Q4 — Minimum Window Substring (Basic idea)
Example:


s = "ADOBECODEBANC"
t = "ABC"
Output


"BANC"
(Sliding window advanced question)

Step-by-Step Idea
Step 1 — Expand Window
Move right pointer until the window contains A, B, C.

Example window


ADOBEC
Now we have:


A ✔
B ✔
C ✔

```python
from collections import Counter

def minWindow(s, t):

    if not s or not t:
        return ""

    countT = Counter(t)
    window = {}

    have = 0
    need = len(countT)

    res = [-1, -1]
    resLen = float("infinity")

    left = 0

    for right in range(len(s)):
        c = s[right]
        window[c] = window.get(c, 0) + 1

        if c in countT and window[c] == countT[c]:
            have += 1

        while have == need:

            if (right - left + 1) < resLen:
                res = [left, right]
                resLen = right - left + 1

            window[s[left]] -= 1

            if s[left] in countT and window[s[left]] < countT[s[left]]:
                have -= 1

            left += 1

    l, r = res
    return s[l:r+1] if resLen != float("infinity") else ""
print(minWindow("ADOBECODEBANC","ABC"))
```

✅ Q5 — Trapping Rain Water
Example:
Input:[0,1,0,2,1,0,1,3,2,1,2,1]Output:6

(Very famous interview problem)

```python
def rainwater(arr):
    if not arr:
        return []
    left=0
    right=len(arr)-1
    left_max=0
    right_max=0
    water=0
    while left<right:
        if arr[left]<arr[right]:
            if arr[left]>=left_max:
                left_max=arr[left]
            else:
                water+=left_max-arr[left]
            left+=1


        else:
            if arr[right]>=right_max:
                right_max=arr[right]
            else:
                water+=right_max-arr[right]
            right-=1
    return water
arr=[0,1,0,2,1,0,1,3,2,1,2,1]
print(rainwater(arr))
        
            
```
